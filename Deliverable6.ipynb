```python
# Step 1:Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Step 2: Loading iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Step 3: Dropping the "species" column
iris.drop(labels = 'species', axis = 1, inplace = True)
iris.head()
```

```python
#Step 4: Adding columns
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
# Step 5: Summary Data Frame
mean = [np.mean(iris['sepal_length']), np.mean(iris['sepal_width']), np.mean(iris['petal_length']),
        np.mean(iris['petal_width']), np.mean(iris['sepal_sum']), np.mean(iris['petal_sum'])]

std = [np.std(iris['sepal_length']), np.std(iris['sepal_width']), np.std(iris['petal_length']),
        np.std(iris['petal_width']), np.std(iris['sepal_sum']), np.std(iris['petal_sum'])]

min1 = [np.min(iris['sepal_length']), np.min(iris['sepal_width']), np.min(iris['petal_length']),
        np.min(iris['petal_width']), np.min(iris['sepal_sum']), np.min(iris['petal_sum'])]

max1 = [np.max(iris['sepal_length']), np.max(iris['sepal_width']), np.max(iris['petal_length']),
        np.max(iris['petal_width']), np.max(iris['sepal_sum']), np.max(iris['petal_sum'])]

summary_dict = {'mean':mean, 'std':std, 'min':min1, 'max':max1}

summary_df = pd.DataFrame(data = summary_dict)

summary_df
```
